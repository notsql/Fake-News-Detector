## Importing
- Importing necessary library
- Importing data

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
```

Combines the dataset

```python
true_news = pd.read_csv('True.csv')
fake_news = pd.read_csv('Fake.csv')

true_news = true_news.merge(pd.DataFrame({'fake': np.full(true_news.shape[0], 0)}), right_index=True, left_index=True)
fake_news = fake_news.merge(pd.DataFrame({'fake': np.full(true_news.shape[0], 1)}), right_index=True, left_index=True)

news_data = pd.concat([true_news, fake_news])
```

## Visualizations

```python
plt.figure(figsize=(13,8))
sb.set_theme(style='whitegrid')

plt.title('Count by Subject')

sb.countplot(data=news_data, x='subject', hue='fake', palette=['#4586FF', '#FBBC05'])

plt.show()
```

# Feature Engineering

Since the there each type of news have their distinct subject, we will be dropping subject.

```python
news_data = news_data.drop(columns=['subject'])
news_data.to_csv('news_data.csv')
```
